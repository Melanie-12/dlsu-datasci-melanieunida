```python
#Exercise3, to change the value of RAD, if 1 to 4 then 1; if 5 to 8 then 2; otherwise of the two conditions then 3

import pandas as pd
import numpy as np

df3 = df.copy()
if "RAD_NEW_COLUMN" not in df3:
    df3["RAD_NEW_COLUMN"] = ""

for index, row in df3.iterrows():
    rad = row['RAD']
    val = ''
    if rad  == 1:
        val = 1
    elif rad == 2:
        val = 1
    elif rad ==3:
        val = 1
    elif rad ==4:
        val = 1
    elif rad ==5:
        val = 2
    elif rad ==6:
        val = 2
    elif rad ==7:
        val =2
    elif rad ==8:
        val =2
else: 3
df3.at[index, 'RAD_NEW_COLUMN'] = val

print(df3)
```
